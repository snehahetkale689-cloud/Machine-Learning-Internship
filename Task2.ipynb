**Step 1 : Import Libraries**

```python
import pandas as pd
```

**Step 2 : Load Dataset**

```python
df = pd.read_csv("Dataset .csv")
df.head()
```

**Step 3 : Handling Missing Values**

```python
df["Cuisines"]=df["Cuisines"].fillna("Unkown")
```

```python
df["Cuisines"]
```

**Step 4 : Creating the recommendation Function**

```python
def recommend_restaurants(cuisine,price_range):
    recommendations = df[
        (df["Cuisines"].str.contains(cuisine,case = False, na=False))&(df["Price range"] == price_range)]
    return recommendations[[
    "Restaurant Name",
    "Cuisines",
    "Average Cost for two",
    "Aggregate rating"
    ]]
```

```python
recommend_restaurants("Italian",2)
```

**Step 5 : Recommend the Highest-Rating Restaurants First**

```python
def recommend_restaurants(cuisine,price_range):
    recommendations=df[
        (df["Cuisines"].str.contains(cuisine,case=False,na=False))&(df["Price range"] == price_range)]
    recommendations=recommendations.sort_values(by="Aggregate rating",ascending=False)
    return recommendations[[
        "Restaurant Name",
        "Cuisines",
        "Average Cost for two",
        "Aggregate rating"
    ]]
        
```

```python
recommend_restaurants("Italian",2)
```

**Step 6 : Recommend Restaurants based on user preferences**

```python
cuisine= input("Enter preferred cuisine:")
price_range = int(input("Enter price range (1-4): "))

recommend_restaurants(cuisine,price_range)
```

**Step 7 : Conclusion:**

A restaurant recommandation system was developed using Python and Pandas.The system recommends restaurants based on the user's preferred cuisine and 
price range.The recommendations are sorted by aggregate rating so that the highest-rated restaurants are shown first.
